# Colab 模板：在 GPU 上运行本项目

目标：在 Google Colab（GPU）中运行本地项目代码（例如 `run_lstm2_pipeline.py`）。

你有三种推荐方式把代码放到 Colab：
1) GitHub（最推荐，最干净）
2) Google Drive（适合有数据/模型文件要同步）
3) 直接上传 zip（最简单但不利于版本管理）


## 0) 运行时设置

在 Colab 菜单：`Runtime -> Change runtime type -> Hardware accelerator -> GPU`。


````python
# 检查 GPU
import torch
print('torch:', torch.__version__)
print('cuda available:', torch.cuda.is_available())
if torch.cuda.is_available():
    print('gpu:', torch.cuda.get_device_name(0))
````

## 1A) 方式一：GitHub（推荐）

把你的项目推到 GitHub 后，在 Colab 里：

```bash

!pip -q install -r requirements.txt
```

> 这种方式最好维护版本、也方便你后续做 Transformer 实验。


## 1B) 方式二：Google Drive（适合存数据/模型）


## 1C) 方式三：上传 zip

在左侧 Files 面板上传 zip 后：

```bash
!unzip -q <YOUR_ZIP>.zip -d /content/fyp
%cd /content/fyp
!pip -q install -r requirements.txt
```
