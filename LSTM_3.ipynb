```python
# %%
# 完整的改进版LSTM策略 - 解决上涨市场亏损问题
import warnings
warnings.filterwarnings("ignore")

import os, math, random, copy
from dataclasses import dataclass
from typing import Dict, List, Tuple, Optional

import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

plt.style.use('seaborn-v0_8-whitegrid') if 'seaborn-v0_8-whitegrid' in plt.style.available else plt.style.use('ggplot')
sns.set_palette("Set2")

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from torch.optim.lr_scheduler import ReduceLROnPlateau

# %% 配置
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
    device = torch.device("cuda")
else:
    device = torch.device("cpu")
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
print("Device:", device)

# %% 改进的配置类
@dataclass
class ImprovedLabelingConfig:
    """改进的标签策略"""
    method: str = "quantile"  # "quantile" or "dynamic"
    up_quantile: float = 0.65  # 改为更平衡的分位数
    down_quantile: float = 0.35
    min_return_threshold: float = 0.005  # 最小收益阈值 0.5%
    use_trend_filter: bool = True  # 使用趋势过滤
    trend_period: int = 20

@dataclass
class ImprovedTrainConfig:
    """改进的训练配置"""
    epochs: int = 150
    batch_size: int = 64
    patience: int = 20
    lr: float = 5e-4  # 降低学习率
    weight_decay: float = 1e-3  # 增加正则化
    lr_factor: float = 0.5
    lr_patience: int = 10
    lr_min: float = 1e-6
    grad_clip: float = 1.0
    
    hidden_size: int = 128  # 增加隐藏层
    num_layers: int = 3  # 增加层数
    dropout: float = 0.3
    attention_heads: int = 4  # 添加attention
    
    use_focal_loss: bool = True  # 使用focal loss
    focal_alpha: float = 0.75
    focal_gamma: float = 2.0
    
    verbose: bool = True

@dataclass
class ImprovedPipelineConfig:
    """改进的流程配置"""
    ticker: str = "2800.HK"
    period: str = "3y"
    interval: str = "1d"
    lookback: int = 40  # 增加lookback
    
    train_window: int = 300  # 增加训练窗口
    val_size: int = 30
    test_size: int = 21
    step_size: int = 10  # 减小步长，增加折数
    
    horizon: int = 1
    
    # 动态仓位管理
    use_dynamic_position: bool = True
    min_confidence: float = 0.55  # 降低阈值
    max_position_size: float = 1.0
    
    # 风险控制
    stop_loss_pct: float = 0.02  # 2% 止损
    max_drawdown_pct: float = 0.10  # 10% 最大回撤控制
    
    # 回测
    transaction_cost_bp: float = 2.0
    backtest_days: int = 252

LABEL_CFG = ImprovedLabelingConfig()
TRAIN_CFG = ImprovedTrainConfig()
PIPE_CFG = ImprovedPipelineConfig()

# %% 数据获取（保持不变）
def fetch_prices(ticker: str, period: str = "3y", interval: str = "1d") -> pd.DataFrame:
    tk = yf.Ticker(ticker)
    df = tk.history(period=period, interval=interval, auto_adjust=True)
    if df is None or df.empty:
        end = pd.Timestamp.today().normalize()
        start = end - pd.Timedelta(days=5 * 370)
        df = tk.history(start=start, end=end, interval=interval, auto_adjust=True)
    if df is None or df.empty:
        raise RuntimeError(f"Cannot fetch data for {ticker}.")
    df.index = pd.to_datetime(df.index).tz_localize(None)
    return df[["Open", "High", "Low", "Close", "Volume"]].copy()

df_raw = fetch_prices(PIPE_CFG.ticker, PIPE_CFG.period, PIPE_CFG.interval)
print(f"Data: {df_raw.index.min().date()} ~ {df_raw.index.max().date()}, N={len(df_raw)}")

# %% 改进的技术指标
def ema(series: pd.Series, span: int):
    return series.ewm(span=span, adjust=False).mean()

def rsi(series: pd.Series, period: int = 14):
    delta = series.diff()
    up = delta.clip(lower=0)
    down = -delta.clip(upper=0)
    roll_up = up.ewm(alpha=1/period, adjust=False).mean()
    roll_down = down.ewm(alpha=1/period, adjust=False).mean().replace(0, np.nan)
    rs = roll_up / roll_down
    return 100 - (100 / (1 + rs))

def macd(close: pd.Series, fast=12, slow=26, signal=9):
    ema_fast = ema(close, fast)
    ema_slow = ema(close, slow)
    macd_line = ema_fast - ema_slow
    signal_line = ema(macd_line, signal)
    return macd_line, signal_line, macd_line - signal_line

def bollinger_bands(close, period=20, num_std=2):
    mid = close.rolling(period).mean()
    sd = close.rolling(period).std()
    upper = mid + num_std * sd
    lower = mid - num_std * sd
    bandwidth = (upper - lower) / mid
    percent_b = (close - lower) / (upper - lower)
    return mid, upper, lower, bandwidth, percent_b

def atr(high, low, close, period=14):
    """Average True Range"""
    tr1 = high - low
    tr2 = abs(high - close.shift())
    tr3 = abs(low - close.shift())
    tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)
    return tr.rolling(period).mean()

def adx(high, low, close, period=14):
    """Average Directional Index"""
    plus_dm = high.diff()
    minus_dm = -low.diff()
    plus_dm[plus_dm < 0] = 0
    minus_dm[minus_dm < 0] = 0
    
    tr = atr(high, low, close, period)
    plus_di = 100 * (plus_dm.ewm(alpha=1/period).mean() / tr)
    minus_di = 100 * (minus_dm.ewm(alpha=1/period).mean() / tr)
    
    dx = 100 * abs(plus_di - minus_di) / (plus_di + minus_di)
    adx = dx.ewm(alpha=1/period).mean()
    return adx

# %% 大幅改进的特征工程
def build_improved_features(df: pd.DataFrame, horizon: int = 1) -> Tuple[pd.DataFrame, List[str]]:
    """构建更丰富的特征"""
    c = df["Close"].copy()
    h = df["High"].copy()
    l = df["Low"].copy()
    v = df["Volume"].copy()
    
    feat = pd.DataFrame(index=df.index)
    
    # 1. 收益率特征（多时间尺度）
    for period in [1, 3, 5, 10, 20]:
        feat[f"ret_{period}d"] = np.log(c / c.shift(period))
    
    # 2. 波动率特征（已实现波动率）
    feat["log_ret"] = np.log(c / c.shift(1))
    for period in [5, 10, 20]:
        feat[f"realized_vol_{period}d"] = feat["log_ret"].rolling(period).std() * np.sqrt(252)
    
    # 3. 趋势特征（多周期均线）
    for period in [5, 10, 20, 50, 100]:
        feat[f"sma_{period}"] = c.rolling(period).mean()
        feat[f"price_to_sma_{period}"] = c / feat[f"sma_{period}"] - 1  # 价格相对位置
    
    # 4. 动量指标
    feat["rsi_14"] = rsi(c, 14)
    feat["rsi_7"] = rsi(c, 7)
    
    # MACD
    macd_line, signal_line, macd_hist = macd(c)
    feat["macd"] = macd_line
    feat["macd_signal"] = signal_line
    feat["macd_hist"] = macd_hist
    
    # 5. Bollinger Bands
    bb_mid, bb_up, bb_lo, bb_bw, bb_pb = bollinger_bands(c, 20, 2)
    feat["bb_percent_b"] = bb_pb
    feat["bb_bandwidth"] = bb_bw
    feat["bb_squeeze"] = (bb_up - bb_lo) / c  # 归一化
    
    # 6. 趋势强度（ADX）
    feat["adx_14"] = adx(h, l, c, 14)
    
    # 7. 波动率指标（ATR）
    feat["atr_14"] = atr(h, l, c, 14)
    feat["atr_ratio"] = feat["atr_14"] / c  # 相对ATR
    
    # 8. 成交量特征
    feat["volume_sma_20"] = v.rolling(20).mean()
    feat["volume_ratio"] = v / feat["volume_sma_20"]
    feat["volume_change"] = v.pct_change()
    
    # 9. 价格行为特征
    feat["high_low_ratio"] = (h - l) / c
    feat["close_position"] = (c - l) / (h - l)  # 收盘在当日范围的位置
    
    # 10. 市场regime特征（趋势vs震荡）
    # 使用20日收益的标准差/均值比率
    rolling_ret = feat["ret_1d"].rolling(20)
    feat["regime_trend_vol"] = rolling_ret.std() / (abs(rolling_ret.mean()) + 1e-6)
    
    # 11. 动量强度
    for period in [5, 10, 20]:
        feat[f"momentum_{period}"] = c / c.shift(period) - 1
    
    # 12. 布林带压力指标
    feat["bb_pressure"] = (c - bb_lo) / (bb_up - bb_lo) - 0.5  # centered
    
    # 目标变量
    feat["next_return"] = np.log(c.shift(-horizon) / c)
    
    # 清理
    feat = feat.replace([np.inf, -np.inf], np.nan)
    data = feat.dropna().copy()
    
    # 特征列（排除目标和中间变量）
    exclude_cols = ["next_return", "log_ret", "volume_sma_20"]
    feature_cols = [col for col in data.columns if col not in exclude_cols]
    
    print(f"✓ Features ready: {len(data)} samples, {len(feature_cols)} features")
    return data, feature_cols

feat_df_all, feature_cols_all = build_improved_features(df_raw, horizon=PIPE_CFG.horizon)
print(f"Feature columns ({len(feature_cols_all)}):")
print(feature_cols_all[:10], "...")

# %% 改进的标签策略
CLASS_ID_DOWN = 0
CLASS_ID_NEUTRAL = 1
CLASS_ID_UP = 2
NUM_CLASSES = 3

def apply_improved_labeling(feat_df: pd.DataFrame, label_cfg: ImprovedLabelingConfig) -> pd.Series:
    """改进的标签策略：quantile + 趋势过滤"""
    ret = feat_df["next_return"]
    
    if label_cfg.method == "quantile":
        window = 60
        upper_thresh = ret.rolling(window, min_periods=20).quantile(label_cfg.up_quantile)
        lower_thresh = ret.rolling(window, min_periods=20).quantile(label_cfg.down_quantile)
        upper_thresh = upper_thresh.fillna(method='ffill').fillna(ret.quantile(label_cfg.up_quantile))
        lower_thresh = lower_thresh.fillna(method='ffill').fillna(ret.quantile(label_cfg.down_quantile))
    else:
        # 使用已有的波动率特征
        vol = feat_df["realized_vol_20d"] / np.sqrt(252)
        upper_thresh = 0.8 * vol
        lower_thresh = -0.8 * vol
    
    upper_thresh = upper_thresh.clip(lower=label_cfg.min_return_threshold)
    lower_thresh = lower_thresh.clip(upper=-label_cfg.min_return_threshold)
    
    labels = pd.Series(CLASS_ID_NEUTRAL, index=feat_df.index, dtype=int)
    labels[ret >= upper_thresh] = CLASS_ID_UP
    labels[ret <= lower_thresh] = CLASS_ID_DOWN
    
    if label_cfg.use_trend_filter:
        # ✅ 关键修复：使用已计算的SMA
        sma_short = feat_df["sma_5"]
        sma_long = feat_df["sma_20"]
        trend_down = sma_short < sma_long
        trend_up = sma_short > sma_long
        
        labels[(labels == CLASS_ID_UP) & trend_down] = CLASS_ID_NEUTRAL
        labels[(labels == CLASS_ID_DOWN) & trend_up] = CLASS_ID_NEUTRAL
    
    return labels

feat_df_labeled = feat_df_all.copy()
feat_df_labeled["target_class"] = apply_improved_labeling(feat_df_labeled, LABEL_CFG)

class_counts = feat_df_labeled["target_class"].value_counts().sort_index()
print(f"\n✓ Label distribution: {class_counts.to_dict()}")
print(f"  Up%={class_counts[2]/len(feat_df_labeled)*100:.1f}%, "
      f"Down%={class_counts[0]/len(feat_df_labeled)*100:.1f}%, "
      f"Neutral%={class_counts[1]/len(feat_df_labeled)*100:.1f}%")

# %% 序列构建（保持不变）
def make_sequences(
    feat_df: pd.DataFrame,
    feature_cols: List[str],
    lookback: int,
    label_cols: List[str]
) -> Tuple[np.ndarray, Dict[str, np.ndarray], pd.DatetimeIndex]:
    X_2d = feat_df[feature_cols].astype(np.float32).values
    dates = feat_df.index
    N = len(feat_df)
    if N < lookback:
        raise ValueError(f"N={N} < lookback={lookback}")
    
    X_list, labels_dict, idx_list = [], {col: [] for col in label_cols}, []
    
    for end in range(lookback - 1, N):
        start = end - lookback + 1
        X_list.append(X_2d[start:end + 1])
        idx_list.append(dates[end])
        for col in label_cols:
            labels_dict[col].append(feat_df.iloc[end][col])
    
    X_seq = np.stack(X_list).astype(np.float32)
    labels_dict = {col: np.asarray(vals) for col, vals in labels_dict.items()}
    return X_seq, labels_dict, pd.DatetimeIndex(idx_list)

def fit_scaler_3d(X_train: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    flat = X_train.reshape(-1, X_train.shape[-1])
    mean = flat.mean(axis=0)
    std = flat.std(axis=0)
    std = np.where(std < 1e-12, 1e-12, std)
    return mean.astype(np.float32), std.astype(np.float32)

def transform_3d(X: np.ndarray, mean: np.ndarray, std: np.ndarray) -> np.ndarray:
    return (X - mean[None, None, :]) / std[None, None, :]

# %% 改进的模型：Attention LSTM
class MultiHeadAttention(nn.Module):
    """多头注意力机制"""
    def __init__(self, hidden_size: int, num_heads: int = 4, dropout: float = 0.1):
        super().__init__()
        assert hidden_size % num_heads == 0
        self.hidden_size = hidden_size
        self.num_heads = num_heads
        self.head_dim = hidden_size // num_heads
        
        self.q_linear = nn.Linear(hidden_size, hidden_size)
        self.k_linear = nn.Linear(hidden_size, hidden_size)
        self.v_linear = nn.Linear(hidden_size, hidden_size)
        self.out = nn.Linear(hidden_size, hidden_size)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        B, T, _ = x.shape
        
        # Linear projections
        q = self.q_linear(x).view(B, T, self.num_heads, self.head_dim).transpose(1, 2)
        k = self.k_linear(x).view(B, T, self.num_heads, self.head_dim).transpose(1, 2)
        v = self.v_linear(x).view(B, T, self.num_heads, self.head_dim).transpose(1, 2)
        
        # Attention
        scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(self.head_dim)
        attn = torch.softmax(scores, dim=-1)
        attn = self.dropout(attn)
        
        # Apply attention to values
        context = torch.matmul(attn, v)
        context = context.transpose(1, 2).contiguous().view(B, T, self.hidden_size)
        
        return self.out(context)

class AttentionLSTM(nn.Module):
    """带注意力机制的LSTM"""
    def __init__(
        self,
        input_size: int,
        hidden_size: int = 128,
        num_layers: int = 3,
        num_heads: int = 4,
        dropout: float = 0.3
    ):
        super().__init__()
        self.hidden_size = hidden_size
        
        # Input projection
        self.input_proj = nn.Linear(input_size, hidden_size)
        
        # LSTM layers
        self.lstm = nn.LSTM(
            input_size=hidden_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0
        )
        
        # Multi-head attention
        self.attention = MultiHeadAttention(hidden_size, num_heads, dropout)
        
        # Layer normalization
        self.ln1 = nn.LayerNorm(hidden_size)
        self.ln2 = nn.LayerNorm(hidden_size)
        
        # Feedforward
        self.ff = nn.Sequential(
            nn.Linear(hidden_size, hidden_size * 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_size * 2, hidden_size),
            nn.Dropout(dropout)
        )
        
        # Output
        self.dropout = nn.Dropout(dropout)
        self.fc_out = nn.Linear(hidden_size, NUM_CLASSES)
        
    def forward(self, x):
        # Input projection
        x = self.input_proj(x)
        
        # LSTM
        lstm_out, _ = self.lstm(x)
        
        # Attention with residual
        attn_out = self.attention(lstm_out)
        x = self.ln1(lstm_out + attn_out)
        
        # Feedforward with residual
        ff_out = self.ff(x)
        x = self.ln2(x + ff_out)
        
        # Take last timestep
        h = x[:, -1, :]
        h = self.dropout(h)
        
        return self.fc_out(h)

# %% Focal Loss（更好处理类别不平衡）
class FocalLoss(nn.Module):
    """Focal Loss for addressing class imbalance"""
    def __init__(self, alpha=0.75, gamma=2.0, ignore_index=-100):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.ignore_index = ignore_index
        
    def forward(self, inputs, targets):
        # Filter out ignored indices
        if self.ignore_index is not None:
            mask = targets != self.ignore_index
            if mask.sum() == 0:
                return torch.zeros((), device=inputs.device)
            inputs = inputs[mask]
            targets = targets[mask]
        
        ce_loss = nn.functional.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * ce_loss
        return focal_loss.mean()

def masked_loss(
    logits: torch.Tensor,
    target: torch.Tensor,
    use_focal: bool = True,
    focal_alpha: float = 0.75,
    focal_gamma: float = 2.0,
    neutral_class_id: int = CLASS_ID_NEUTRAL
) -> torch.Tensor:
    """改进的损失函数"""
    # 只计算Up/Down的损失
    mask = target != neutral_class_id
    if mask.sum() == 0:
        return torch.zeros((), device=logits.device)
    
    if use_focal:
        # 标记Neutral为ignore
        target_masked = target.clone()
        target_masked[~mask] = -100
        criterion = FocalLoss(alpha=focal_alpha, gamma=focal_gamma, ignore_index=-100)
        return criterion(logits, target_masked)
    else:
        logits_sel = logits[mask]
        target_sel = target[mask]
        return nn.functional.cross_entropy(logits_sel, target_sel)

# %% 训练和预测函数
def make_loader(X: np.ndarray, y: np.ndarray, batch_size: int, shuffle: bool) -> DataLoader:
    X_t = torch.from_numpy(X.astype(np.float32))
    y_t = torch.from_numpy(y.astype(np.int64))
    ds = TensorDataset(X_t, y_t)
    return DataLoader(ds, batch_size=batch_size, shuffle=shuffle, num_workers=0, pin_memory=torch.cuda.is_available())

def train_improved_model(
    model: nn.Module,
    train_loader: DataLoader,
    val_loader: DataLoader,
    cfg: ImprovedTrainConfig
) -> Tuple[nn.Module, Dict[str, List[float]]]:
    """改进的训练流程"""
    model = model.to(device)
    optimizer = optim.AdamW(model.parameters(), lr=cfg.lr, weight_decay=cfg.weight_decay)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=cfg.lr_factor, patience=cfg.lr_patience, min_lr=cfg.lr_min)
    
    best_state = copy.deepcopy(model.state_dict())
    best_val = float('inf')
    no_improve = 0
    history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
    
    for epoch in range(1, cfg.epochs + 1):
        # Training
        model.train()
        train_loss, train_correct, train_total = 0.0, 0, 0
        
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            logits = model(xb)
            loss = masked_loss(logits, yb, cfg.use_focal_loss, cfg.focal_alpha, cfg.focal_gamma)
            loss.backward()
            if cfg.grad_clip:
                nn.utils.clip_grad_norm_(model.parameters(), cfg.grad_clip)
            optimizer.step()
            
            train_loss += loss.item() * xb.size(0)
            
            # Accuracy (only on Up/Down)
            mask = yb != CLASS_ID_NEUTRAL
            if mask.sum() > 0:
                pred = logits.argmax(dim=1)
                train_correct += (pred[mask] == yb[mask]).sum().item()
                train_total += mask.sum().item()
        
        train_loss = train_loss / len(train_loader.dataset)
        train_acc = train_correct / max(train_total, 1)
        
        # Validation
        model.eval()
        val_loss, val_correct, val_total = 0.0, 0, 0
        
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                logits = model(xb)
                loss = masked_loss(logits, yb, cfg.use_focal_loss, cfg.focal_alpha, cfg.focal_gamma)
                val_loss += loss.item() * xb.size(0)
                
                mask = yb != CLASS_ID_NEUTRAL
                if mask.sum() > 0:
                    pred = logits.argmax(dim=1)
                    val_correct += (pred[mask] == yb[mask]).sum().item()
                    val_total += mask.sum().item()
        
        val_loss = val_loss / len(val_loader.dataset)
        val_acc = val_correct / max(val_total, 1)
        
        scheduler.step(val_loss)
        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)
        
        if cfg.verbose and (epoch == 1 or epoch % 10 == 0):
            print(f"Epoch {epoch:03d} | TrL {train_loss:.4f} VaL {val_loss:.4f} | "
                  f"TrAcc {train_acc:.3f} VaAcc {val_acc:.3f} | LR {optimizer.param_groups[0]['lr']:.2e}")
        
        if val_loss + 1e-9 < best_val:
            best_val = val_loss
            best_state = copy.deepcopy(model.state_dict())
            no_improve = 0
        else:
            no_improve += 1
            if no_improve >= cfg.patience:
                if cfg.verbose:
                    print(f"✓ Early stop at epoch {epoch}, best_val={best_val:.4f}")
                break
    
    model.load_state_dict(best_state)
    return model, history

def predict_proba(model: nn.Module, X: np.ndarray, batch_size: int = 256) -> np.ndarray:
    """预测概率"""
    model = model.to(device)
    model.eval()
    outputs = []
    with torch.no_grad():
        for i in range(0, len(X), batch_size):
            xb = torch.from_numpy(X[i:i+batch_size].astype(np.float32)).to(device)
            logits = model(xb)
            proba = torch.softmax(logits, dim=1)
            outputs.append(proba.cpu().numpy())
    return np.concatenate(outputs, axis=0)

# %% Walk-Forward CV（改进版）
def improved_walk_forward_cv(
    X_seq_all: np.ndarray,
    y_all: np.ndarray,
    seq_index: pd.DatetimeIndex,
    cfg_pipe: ImprovedPipelineConfig,
    cfg_train: ImprovedTrainConfig
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """改进的walk-forward验证"""
    metrics, preds_list = [], []
    N = len(X_seq_all)
    start_index = cfg_pipe.train_window + cfg_pipe.val_size
    fold_id = 0
    
    print(f"\n{'='*60}")
    print(f"Starting Walk-Forward CV")
    print(f"Train window: {cfg_pipe.train_window}, Val: {cfg_pipe.val_size}, Test: {cfg_pipe.test_size}")
    print(f"{'='*60}\n")
    
    for test_start in range(start_index, N - cfg_pipe.test_size + 1, cfg_pipe.step_size):
        train_end = test_start
        val_start = train_end - cfg_pipe.val_size
        train_start = val_start - cfg_pipe.train_window
        test_end = test_start + cfg_pipe.test_size
        
        if train_start < 0:
            continue
        
        # Split data
        X_train = X_seq_all[train_start:val_start]
        y_train = y_all[train_start:val_start]
        X_val = X_seq_all[val_start:train_end]
        y_val = y_all[val_start:train_end]
        X_test = X_seq_all[test_start:test_end]
        y_test = y_all[test_start:test_end]
        test_dates = seq_index[test_start:test_end]
        
        # Check sufficient Up/Down samples
        mask_ud_train = y_train != CLASS_ID_NEUTRAL
        mask_ud_val = y_val != CLASS_ID_NEUTRAL
        if mask_ud_train.sum() < 20 or mask_ud_val.sum() < 5:
            print(f"Fold {fold_id}: Skipped (insufficient Up/Down samples)")
            fold_id += 1
            continue
        
        # Standardize
        mean, std = fit_scaler_3d(X_train)
        X_train_sc = transform_3d(X_train, mean, std)
        X_val_sc = transform_3d(X_val, mean, std)
        X_test_sc = transform_3d(X_test, mean, std)
        
        # Model
        model = AttentionLSTM(
            input_size=X_train_sc.shape[-1],
            hidden_size=cfg_train.hidden_size,
            num_layers=cfg_train.num_layers,
            num_heads=cfg_train.attention_heads,
            dropout=cfg_train.dropout
        )
        
        train_loader = make_loader(X_train_sc, y_train, cfg_train.batch_size, shuffle=True)
        val_loader = make_loader(X_val_sc, y_val, cfg_train.batch_size, shuffle=False)
        
        print(f"\nFold {fold_id}: {test_dates[0].date()} ~ {test_dates[-1].date()}")
        print(f"  Train: {len(X_train)} | Val: {len(X_val)} | Test: {len(X_test)}")
        print(f"  Train Up/Down: {mask_ud_train.sum()} ({mask_ud_train.mean()*100:.1f}%)")
        
        # Train
        model, hist = train_improved_model(model, train_loader, val_loader, cfg_train)
        
        # Predict
        proba_test = predict_proba(model, X_test_sc)
        pred_class = proba_test.argmax(axis=1)
        
        # 应用置信度过滤（可选）
        max_proba = proba_test.max(axis=1)
        if cfg_pipe.use_dynamic_position:
            pred_class[max_proba < cfg_pipe.min_confidence] = CLASS_ID_NEUTRAL
        
        # Metrics
        y_true = y_test
        mask_pred_up = pred_class == CLASS_ID_UP
        mask_pred_down = pred_class == CLASS_ID_DOWN
        mask_pred_ud = mask_pred_up | mask_pred_down
        
        up_acc = accuracy_score(y_true[mask_pred_up], pred_class[mask_pred_up]) if mask_pred_up.any() else np.nan
        down_acc = accuracy_score(y_true[mask_pred_down], pred_class[mask_pred_down]) if mask_pred_down.any() else np.nan
        dir_acc = accuracy_score(y_true[mask_pred_ud], pred_class[mask_pred_ud]) if mask_pred_ud.any() else np.nan
        
        metrics.append({
            "fold": fold_id,
            "test_start": test_dates[0],
            "test_end": test_dates[-1],
            "up_acc": up_acc,
            "down_acc": down_acc,
            "dir_acc": dir_acc,
            "coverage": mask_pred_ud.mean(),
            "num_pred_up": int(mask_pred_up.sum()),
            "num_pred_down": int(mask_pred_down.sum()),
            "test_size": len(X_test),
        })
        
        print(f"  Results: DirAcc={dir_acc:.3f}, UpAcc={up_acc:.3f}, DownAcc={down_acc:.3f}, Cov={mask_pred_ud.mean():.2f}")
        
        fold_pred = pd.DataFrame({
            "fold": fold_id,
            "date": test_dates,
            "actual_class": y_true,
            "pred_class": pred_class,
            "proba_down": proba_test[:, CLASS_ID_DOWN],
            "proba_neutral": proba_test[:, CLASS_ID_NEUTRAL],
            "proba_up": proba_test[:, CLASS_ID_UP],
            "confidence": max_proba,
        })
        preds_list.append(fold_pred)
        
        fold_id += 1
        del model
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
    
    return pd.DataFrame(metrics), pd.concat(preds_list, ignore_index=True) if preds_list else pd.DataFrame()

# %% 准备数据并运行CV
X_seq, label_dict, seq_index = make_sequences(
    feat_df_labeled,
    feature_cols=feature_cols_all,
    lookback=PIPE_CFG.lookback,
    label_cols=["target_class", "next_return"]
)
y_all = label_dict["target_class"].astype(np.int64)

print(f"\n✓ Sequences ready: {len(X_seq)} samples, lookback={PIPE_CFG.lookback}, features={X_seq.shape[-1]}")

cv_metrics, cv_preds = improved_walk_forward_cv(X_seq, y_all, seq_index, PIPE_CFG, TRAIN_CFG)

# %% CV结果分析
print("\n" + "="*60)
print("CV Metrics Summary")
print("="*60)
display(cv_metrics)

if not cv_metrics.empty:
    summary = cv_metrics[["up_acc", "down_acc", "dir_acc", "coverage"]].agg(["mean", "std", "min", "max"])
    print("\n✓ Summary Statistics:")
    display(summary)

# %% 改进的回测：动态仓位 + 风险控制
def improved_backtest(
    cv_preds: pd.DataFrame,
    feat_df: pd.DataFrame,
    cfg: ImprovedPipelineConfig
) -> Tuple[pd.DataFrame, Dict]:
    """改进的回测：动态仓位+止损"""
    if cv_preds.empty:
        raise ValueError("Empty predictions")
    
    # 准备数据
    df = cv_preds.sort_values("date").copy()
    df["date"] = pd.to_datetime(df["date"]).dt.tz_localize(None)
    
    ret_df = feat_df[["next_return"]].reset_index()
    ret_df.columns = ["date", "next_return"]
    ret_df["date"] = pd.to_datetime(ret_df["date"]).dt.tz_localize(None)
    
    bt = df.merge(ret_df, on="date", how="left")
    bt["simple_return"] = np.exp(bt["next_return"].fillna(0)) - 1.0
    
    # 取最近N天
    if len(bt) > cfg.backtest_days:
        bt = bt.iloc[-cfg.backtest_days:].copy()
    
    # 动态仓位计算
    def compute_position_size(row):
        """根据置信度计算仓位"""
        if not cfg.use_dynamic_position:
            # 简单策略
            if row["pred_class"] == CLASS_ID_UP:
                return 1.0
            elif row["pred_class"] == CLASS_ID_DOWN:
                return -1.0
            return 0.0
        
        # 动态仓位
        conf = row["confidence"]
        pred = row["pred_class"]
        
        if conf < cfg.min_confidence:
            return 0.0
        
        # 线性映射：confidence -> position size
        # confidence in [min_conf, 1.0] -> size in [0.3, 1.0]
        size = 0.3 + 0.7 * (conf - cfg.min_confidence) / (1.0 - cfg.min_confidence)
        size = np.clip(size, 0, cfg.max_position_size)
        
        if pred == CLASS_ID_UP:
            return size
        elif pred == CLASS_ID_DOWN:
            return -size
        return 0.0
    
    bt["raw_position"] = bt.apply(compute_position_size, axis=1)
    
    # 止损逻辑
    positions = bt["raw_position"].values.copy()
    cumret = 0.0
    max_cumret = 0.0
    
    for i in range(len(bt)):
        if i > 0:
            # 计算当日收益
            day_ret = positions[i-1] * bt["simple_return"].iloc[i]
            cumret += day_ret
            max_cumret = max(max_cumret, cumret)
            
            # 止损条件
            if positions[i-1] != 0:
                # 单日止损
                if day_ret < -cfg.stop_loss_pct:
                    positions[i] = 0.0
                # 最大回撤止损
                drawdown = (cumret - max_cumret)
                if drawdown < -cfg.max_drawdown_pct:
                    positions[i] = 0.0
    
    bt["position"] = positions
    
    # 计算收益
    bt["strategy_return"] = bt["position"] * bt["simple_return"]
    
    # 交易成本
    position_change = bt["position"].diff().abs()
    position_change.iloc[0] = abs(bt["position"].iloc[0])
    cost = position_change * (cfg.transaction_cost_bp / 10000.0)
    bt["transaction_cost"] = cost
    bt["strategy_after_cost"] = bt["strategy_return"] - cost
    
    # 权益曲线
    bt["strategy_equity"] = (1 + bt["strategy_after_cost"]).cumprod()
    bt["buyhold_equity"] = (1 + bt["simple_return"]).cumprod()
    
    # 统计
    days = len(bt)
    years = max(days / 252, 1/252)
    final_eq = bt["strategy_equity"].iloc[-1]
    bh_eq = bt["buyhold_equity"].iloc[-1]
    
    daily_ret = bt["strategy_after_cost"]
    mean_ret = daily_ret.mean()
    std_ret = daily_ret.std()
    
    # 最大回撤
    cummax = bt["strategy_equity"].cummax()
    dd = (bt["strategy_equity"] / cummax - 1).min()
    
    stats = {
        "days": days,
        "total_return": final_eq - 1.0,
        "buyhold_return": bh_eq - 1.0,
        "annual_return": final_eq ** (1/years) - 1.0,
        "buyhold_annual": bh_eq ** (1/years) - 1.0,
        "annual_vol": std_ret * np.sqrt(252),
        "sharpe": (mean_ret / std_ret) * np.sqrt(252) if std_ret > 1e-8 else np.nan,
        "max_dd": dd,
        "win_rate": (daily_ret[bt["position"] != 0] > 0).mean() if (bt["position"] != 0).any() else np.nan,
        "avg_win": daily_ret[daily_ret > 0].mean() if (daily_ret > 0).any() else 0,
        "avg_loss": daily_ret[daily_ret < 0].mean() if (daily_ret < 0).any() else 0,
        "coverage": (bt["position"] != 0).mean(),
    }
    
    return bt, stats

bt_df, bt_stats = improved_backtest(cv_preds, feat_df_labeled, PIPE_CFG)

print("\n" + "="*60)
print("Backtest Results (Last ~1 Year)")
print("="*60)
for k, v in bt_stats.items():
    if isinstance(v, float):
        print(f"{k:20s}: {v:>10.4f}")
    else:
        print(f"{k:20s}: {v:>10}")

# %% 可视化
def plot_improved_backtest(bt_df: pd.DataFrame):
    """改进的回测可视化"""
    fig = plt.figure(figsize=(14, 10))
    gs = fig.add_gridspec(4, 1, height_ratios=[3, 1, 1, 1], hspace=0.3)
    
    # 权益曲线
    ax1 = fig.add_subplot(gs[0])
    ax1.plot(bt_df["date"], bt_df["strategy_equity"], label="Strategy", linewidth=2, color="tab:blue")
    ax1.plot(bt_df["date"], bt_df["buyhold_equity"], label="Buy & Hold", linewidth=2, linestyle="--", color="tab:orange")
    ax1.set_ylabel("Equity", fontsize=12)
    ax1.set_title("Improved LSTM Strategy - 1Y Backtest", fontsize=14, fontweight="bold")
    ax1.legend(loc="best", fontsize=10)
    ax1.grid(True, alpha=0.3)
    
    # 仓位
    ax2 = fig.add_subplot(gs[1], sharex=ax1)
    ax2.step(bt_df["date"], bt_df["position"], where="post", color="tab:green", linewidth=1.5)
    ax2.axhline(0, color="gray", linestyle="--", alpha=0.5)
    ax2.set_ylabel("Position", fontsize=12)
    ax2.set_ylim(-1.2, 1.2)
    ax2.grid(True, alpha=0.3)
    
    # 置信度
    ax3 = fig.add_subplot(gs[2], sharex=ax1)
    ax3.plot(bt_df["date"], bt_df["confidence"], color="purple", alpha=0.7, label="Confidence")
    ax3.axhline(PIPE_CFG.min_confidence, color="red", linestyle="--", alpha=0.6, label=f"Threshold={PIPE_CFG.min_confidence}")
    ax3.set_ylabel("Confidence", fontsize=12)
    ax3.legend(loc="best", fontsize=9)
    ax3.grid(True, alpha=0.3)
    
    # 每日收益
    ax4 = fig.add_subplot(gs[3], sharex=ax1)
    colors = ['red' if x < 0 else 'green' for x in bt_df["strategy_after_cost"]]
    ax4.bar(bt_df["date"], bt_df["strategy_after_cost"], color=colors, alpha=0.6, width=0.8)
    ax4.axhline(0, color="black", linewidth=0.8)
    ax4.set_ylabel("Daily Return", fontsize=12)
    ax4.set_xlabel("Date", fontsize=12)
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

plot_improved_backtest(bt_df)

# %% 详细分析报告
def detailed_analysis(cv_preds: pd.DataFrame):
    """详细分析预测结果"""
    df = cv_preds.copy()
    
    # 只看有方向信号的
    mask_signal = (df["pred_class"] == CLASS_ID_UP) | (df["pred_class"] == CLASS_ID_DOWN)
    df_signal = df[mask_signal].copy()
    
    if df_signal.empty:
        print("No directional signals found.")
        return
    
    y_true = df_signal["actual_class"]
    y_pred = df_signal["pred_class"]
    
    print("\n" + "="*60)
    print("Detailed Signal Analysis")
    print("="*60)
    
    # 整体准确率
    acc = accuracy_score(y_true, y_pred)
    print(f"\nOverall Directional Accuracy: {acc:.3f}")
    
    # 分类报告
    print("\nClassification Report (Up/Down only):")
    print(classification_report(
        y_true, y_pred,
        labels=[CLASS_ID_DOWN, CLASS_ID_UP],
        target_names=["Down", "Up"],
        digits=3, zero_division=0
    ))
    
    # 混淆矩阵
    cm = confusion_matrix(y_true, y_pred, labels=[CLASS_ID_DOWN, CLASS_ID_UP])
    cm_df = pd.DataFrame(cm, index=["True Down", "True Up"], columns=["Pred Down", "Pred Up"])
    print("\nConfusion Matrix:")
    display(cm_df)
    
    # 按置信度分层分析
    df_signal["conf_bin"] = pd.cut(df_signal["confidence"], bins=[0, 0.6, 0.7, 0.8, 1.0], labels=["Low", "Mid", "High", "VHigh"])
    print("\nAccuracy by Confidence Level:")
    for level in ["Low", "Mid", "High", "VHigh"]:
        sub = df_signal[df_signal["conf_bin"] == level]
        if len(sub) > 0:
            acc_level = accuracy_score(sub["actual_class"], sub["pred_class"])
            print(f"  {level:6s}: {acc_level:.3f} (n={len(sub)})")

detailed_analysis(cv_preds)

print("\n" + "="*60)
print("✓ Analysis Complete!")
print("="*60)
```
